Ciclos en python

```python
opcion = 0

while opcion != 3:
  print("¿que acion deseas realizar?")
  print("1. Saludar")
  print("2. Sumar dos numeros")
  print("3. Salir")

  opcion = int(input("Ingresa tu opcion: "))

  if opcion == 1:
    print("Hola mundo")
  elif opcion == 2:
    num1, num2 = map(int, input("Ingrese dos numeros separados por espacio: ").split())
    print (f"La suma de {num1} y {num2} es {num1+num2}")
  elif opcion == 3:
    print("adios")
  else:
    print("opcion invalida")
```

```python
password = "adivina"
intentos = 0

while intentos <3:
     respuesta = input("ingrse contraseña:")
     intentos += 1
     restantes = 3-intentos
     if respuesta != password and restantes !=0: # reduced indentation by one level
         print(f"contraseña incorrecta.le quedan {restantes} intentos")
     else:
         break

if respuesta == password:
    print("contraseña correcta")
else:
    print("acceso denegado")

```

```python
lista1 = ["manzana", "naranja", "uva", "fresa", "kiwi"]
lista2 = ["sandía", "mango", "kiwi", "fresa", "limón"]

comunes = []
for elemento in lista1:
  if elemento in lista2:
    comunes.append(elemento)
    print("Elementos comunes en ambas listas: ", comunes)

    comunes = [elemento for elemento in lista1 if elemento in lista2]
    print("Elementos comunes en ambas listas: ", comunes)

```
